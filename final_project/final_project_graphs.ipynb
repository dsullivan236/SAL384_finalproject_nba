```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
full_clean = pd.read_csv("full_updated.csv")
full_clean = full_clean.drop(columns = ['Unnamed: 0.4','Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 
                                       'Unnamed: 0','Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'])
full_clean
```

```python
full_clean['acc_nba_ppg_average'] = full_clean.loc[full_clean['College_Conference'] == 'ACC', 'NBA_PPG'].mean()
full_clean['sec_nba_ppg_average'] = full_clean.loc[full_clean['College_Conference'] == 'SEC', 'NBA_PPG'].mean()
full_clean['big12_nba_ppg_average'] = full_clean.loc[full_clean['College_Conference'] == 'Big 12', 'NBA_PPG'].mean()
full_clean['big10_nba_ppg_average'] = full_clean.loc[full_clean['College_Conference'] == 'Big Ten', 'NBA_PPG'].mean()
full_clean['pac12_nba_ppg_average'] = full_clean.loc[full_clean['College_Conference'] == 'Pac 12', 'NBA_PPG'].mean()
full_clean.head()
# Order of highest PPG scorers in the NBA by conference
# pac 12
# sec
# big 12
# acc
# big 10
```

```python
plt.scatter(full_clean.College_Conference, full_clean.NBA_PPG)
plt.title('PPG by College Conference')
plt.xlabel('College_Conference')
plt.ylabel('NBA_PPG')
```

```python
plt.scatter(full_clean.College_PPG, full_clean.NBA_PPG, color = 'r')
plt.title('College_PPG v NBA_PPG')
plt.xlabel('College_PPG')
plt.ylabel('NBA_PPG')
```

```python
plt.hist(full_clean.College_PPG, bins=20, color='green')
plt.title('Histogram of College PPG')
plt.xlabel('College PPG')
plt.ylabel('count')
```

```python
plt.hist(full_clean.NBA_PPG, bins=20, color='purple')
plt.title('Histogram of NBA PPG')
plt.xlabel('NBA PPG')
plt.ylabel('count')
```
